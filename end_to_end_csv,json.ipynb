```python
import pandas as pd

url = "https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv"

# Example 1: Default UTF-8 encoding
df_utf8 = pd.read_csv(url, encoding="utf-8")
print(df_utf8.head())

# Example 2: For demonstration, forcing Latin-1 encoding
df_latin1 = pd.read_csv(url, encoding="latin1")
print(df_latin1.head())
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv"

# Example 1: Parse a single column as dates
df = pd.read_csv(url, parse_dates=["Date"])
print(df.dtypes)   # 'Date' will now be datetime64[ns]

```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv"


# Example 2: Convert the 'Date' column into a custom format string
df2 = pd.read_csv(url, converters={"Date": lambda x: x.replace("-", "/")})
print(df2.head())

# Example 3: Apply a custom transformation (e.g., uppercase text)
# Suppose the CSV had a 'Company' column
df3 = pd.read_csv(url, converters={"Company": lambda x: x.upper()})
```
