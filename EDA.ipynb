```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
#categoricaal data 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(df['Survived'])
df['Survived'].value_counts().plot(kind='bar')
```

```python
#piechart
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
#numerical data
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=50)
```

```python
sns.distplot(df['Age'])#probablity density function
```

```python
#boxplot
sns.boxplot(df['Fare'])
```

```python
df['Age'].min()
```

```python
df['Age'].max()
```

```python
#bivariate aalysis
#multivariate analysis
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'],hue=tips['sex'],style= tips['smoker'])##multivariate analysis
```

```python
sns.barplot(x="Pclass", y="Fare", hue="Sex", data=df)
```

```python
sns.boxplot(x='Sex',y='Age',hue='Survived',data=df)
```

```python
sns.distplot(df[df['Survived']==0]['Age'],hist=False)##jo survive nahi kiye unke age ka hai,blue wala
sns.distplot(df[df['Survived']==1]['Age'],hist=False)##jo survive  kiye unke age ka hai,yellow wala
```

```python
#categorical-categrical
sns.heatmap(pd.crosstab(df['Pclass'],df['Survived']))
```

```python
df['Survived'] = pd.to_numeric(df['Survived'], errors='coerce')
(df.groupby('Pclass')['Survived'].mean() * 100)
#percentage of people survived
```

```python
#pairplot(numerical-numerical)
sns.pairplot(df,hue='Survived')##showing  rations between numerical veriables
```

```python
#numerical-numerical
#if time ,date,number etc are on x axis then use lineplot
k=tips.groupby('day')['size'].sum().reset_index()
```

```python
sns.lineplot(x='day', y='size', data=k)
```

```python
sns.heatmap(tips.pivot_table(values='total_bill',index='day',columns='time'))
```
